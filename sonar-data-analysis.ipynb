**1. Reading the CSV File**

```python
sonar = pd.read_excel('../input/sonardata/sonar.xlsx', header = None)
X = sonar.iloc[:,0:60].values
y = sonar.iloc[:,60].values
sonar
```

**2. Assessing the Data**

```python
sonar.info()
```

**3. Train-Test Split**

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**4. The Model**

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
classifier = Sequential()

classifier.add(Dense(units = 45, activation = 'relu', kernel_initializer = 'uniform', input_dim = 60))
classifier.add(Dropout(rate = 0.1))

classifier.add(Dense(units = 45, activation = 'relu', kernel_initializer = 'uniform'))
classifier.add(Dropout(rate = 0.1))

classifier.add(Dense(units = 1, activation = 'sigmoid', kernel_initializer = 'uniform'))
```

**5. Using the model on the data**

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 35, epochs = 200)
```

```python
y_pred = classifier.predict(X_test) > 0.6
y_pred
```

**6. Scoring**

For scoring, the following have been used:
* 1) Confusion matrix whose i-th row and j-th column entry indicates the number of samples with true label being i-th class and prediced label being j-th class.
* 2) Precision is the ability of the classifier not to label as positive a sample that is negative, and recall is the ability of the classifier to find all the positive samples.
* 3) The F-measure can be interpreted as a weighted harmonic mean of the precision and recall. An F1 score reaches its best value at 1 and its worst score at 0.
* 4) The precision_recall_curve computes a precision-recall curve from the ground truth label and a score given by the classifier by varying a decision threshold.
* 5) The balanced_accuracy_score function computes the balanced accuracy, which avoids inflated performance estimates on imbalanced datasets.

Confusion Matrix:

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sn
norm = mpl.colors.Normalize()
con = confusion_matrix(y_test, y_pred)
sn.heatmap(con, annot=True)
```

```python
con
```

Precison, Recall, F1 Score:

```python
from sklearn.metrics import precision_recall_curve
f1 = metrics.f1_score(y_test, y_pred)
aps = metrics.average_precision_score(y_test, y_pred)
prec = metrics.precision_recall_curve(y_test,y_pred)
print('Precision Recall Curve: ',prec)
print('F1 Score: ',f1)
print('Average Precision Score: ',aps)
```

Accuracy:

```python
acc = metrics.balanced_accuracy_score(y_test, y_pred)
```

```python
print('Balanced Accuracy Score: ',acc)
```
